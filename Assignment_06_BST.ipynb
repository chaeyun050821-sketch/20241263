## 문제 01
다음 이진 검색 트리에 대해 답하시오.

```
          40
         /  \
        10   40
       /  \    \
      7   20   70
          /
         17
```

### 01-1) 이 이진 검색 트리의 리프 노드는 총 몇 개인가?

**풀이:**  
리프 노드란 자식 노드가 없는 노드이다.  
위 트리에서 자식이 없는 노드: **7, 17, 70**

**정답: 3개**

### 01-2) 현재 높이 4를 유지하면서 키를 추가한다면 가능한 리프 노드의 최대 개수는 몇 개인가?

**풀이:**  
높이 4인 이진 트리에서 리프 노드의 최대 개수는 높이 4인 레벨(4번째 레벨)에 노드를 최대한 채울 때 발생한다.  
높이 h인 포화 이진 트리의 리프 노드 수 = 2^(h-1)  
높이 4이면 리프 노드 최대 = 2^(4-1) = **8개**

현재 트리에서 높이 4 레벨에 노드를 추가할 수 있는 위치:  
- 7의 왼쪽, 오른쪽 자식
- 17의 오른쪽 자식 (왼쪽 자식에 17이 있으므로)
- 70의 왼쪽, 오른쪽 자식
- 40(오른쪽 자식)의 왼쪽 자식

높이 4를 유지하면서 레벨 4에 노드를 최대한 추가하면 리프 노드 최대 = **8개**

**정답: 8개**

### 01-3) 높이 4인 이진 검색 트리가 포화 이진 트리 모양이 되면 리프 노드와 내부 노드의 수는 몇 개나 차이가 나는가?

**풀이:**  
높이 h인 포화 이진 트리의 전체 노드 수 = 2^h - 1  
높이 4이면 전체 노드 수 = 2^4 - 1 = 15개  

- 리프 노드 수 = 2^(h-1) = 2^3 = 8개  
- 내부 노드 수 = 15 - 8 = 7개  
- 차이 = 8 - 7 = **1개**

**정답: 1개 (리프 노드가 내부 노드보다 1개 많다)**

### 01-4) 위 3)의 성질은 높이가 h인 모든 포화 이진 트리에 대해 성립하는가?

**풀이:**  
높이 h인 포화 이진 트리에서:  
- 전체 노드 수 = 2^h - 1  
- 리프 노드 수 = 2^(h-1)  
- 내부 노드 수 = 2^h - 1 - 2^(h-1) = 2^(h-1) - 1  
- 차이 = 2^(h-1) - (2^(h-1) - 1) = **1**

**정답: 예, 높이 h인 모든 포화 이진 트리에서 리프 노드 수는 내부 노드 수보다 항상 정확히 1개 많다.**

## 문제 02
임의의 완전 이진 트리에 대한 다음 물음에 답하시오.

### 02-1) 트리의 총 노드 수가 n이면 리프 노드의 수는 몇 개인가?

**풀이:**  
완전 이진 트리에서 총 노드 수가 n일 때:

- n이 짝수인 경우: 리프 노드 수 = n / 2
- n이 홀수인 경우: 리프 노드 수 = (n + 1) / 2

이를 합쳐서 표현하면: **⌈n/2⌉** (n을 2로 나눈 올림값)

**정답: ⌈n/2⌉개**

### 02-2) 위 1)에서 리프 노드의 수와 내부 노드의 수는 어떤 관계인가?

**풀이:**  
- 리프 노드 수 = ⌈n/2⌉  
- 내부 노드 수 = n - ⌈n/2⌉ = ⌊n/2⌋

따라서:
- n이 홀수: 리프 = (n+1)/2, 내부 = (n-1)/2 → 리프 노드가 내부 노드보다 **1개 많다**
- n이 짝수: 리프 = n/2, 내부 = n/2 → 리프 노드 수와 내부 노드 수가 **같다**

**정답: 리프 노드 수(⌈n/2⌉)는 내부 노드 수(⌊n/2⌋)보다 같거나 1개 많다.**

## 문제 03
다음 이진 검색 트리에 대해 답하시오.

```
        5
       / \
      3   8
     /   / \
    3   7   9
       /
      6
```

### 03-1) 전위 순회 (Preorder: 루트 → 왼쪽 → 오른쪽)

**풀이:**  
- 5 방문
- 왼쪽 서브트리: 3 → 3
- 오른쪽 서브트리: 8 → 7 → 6 → 9

**정답: 5 → 3 → 3 → 8 → 7 → 6 → 9**

### 03-2) 중위 순회 (Inorder: 왼쪽 → 루트 → 오른쪽)

**풀이:**  
- 왼쪽 서브트리: 3 → 3
- 루트: 5
- 오른쪽 서브트리: 6 → 7 → 8 → 9

**정답: 3 → 3 → 5 → 6 → 7 → 8 → 9**

### 03-3) 후위 순회 (Postorder: 왼쪽 → 오른쪽 → 루트)

**풀이:**  
- 왼쪽 서브트리: 3 → 3
- 오른쪽 서브트리: 6 → 7 → 9 → 8
- 루트: 5

**정답: 3 → 3 → 6 → 7 → 9 → 8 → 5**

### 03-4) 중위 순회는 모든 이진 검색 트리에 대해 항상 키를 정렬된 순서로 방문한다는 사실을 증명하시오.

**증명 (수학적 귀납법):**

**[이진 검색 트리 성질]**  
임의의 노드 v에 대해:
- v의 왼쪽 서브트리의 모든 키 < v의 키
- v의 오른쪽 서브트리의 모든 키 > v의 키

**[귀납적 증명]**

- **기저 사례:** 노드가 1개인 트리에서 중위 순회는 그 노드 하나만 방문하므로 정렬된 순서이다.

- **귀납 가정:** 노드 수가 k보다 작은 모든 이진 검색 트리에 대해 중위 순회가 정렬된 순서를 출력한다고 가정한다.

- **귀납 단계:** 노드 수가 k인 이진 검색 트리 T의 루트를 r, 왼쪽 서브트리를 L, 오른쪽 서브트리를 R이라 하면:  
  1. L의 노드 수 < k이므로 귀납 가정에 의해 중위 순회(L)는 정렬된 순서 출력  
  2. R의 노드 수 < k이므로 귀납 가정에 의해 중위 순회(R)는 정렬된 순서 출력  
  3. BST 성질에 의해: L의 모든 키 < r의 키 < R의 모든 키  
  4. 중위 순회는 L → r → R 순으로 방문하므로, 전체 방문 순서도 오름차순 정렬 순서이다.

**결론: 귀납법에 의해, 모든 이진 검색 트리에서 중위 순회는 키를 정렬된(오름차순) 순서로 방문한다.**

## 문제 09
n개의 원소로 구성되는 이진 검색 트리의 가능한 최대 높이와 최소 높이는 어떻게 되는가?

**풀이:**

### 최대 높이
이진 검색 트리가 **편향 트리(skewed tree)**가 될 때 높이가 최대이다.  
각 노드가 자식을 하나만 가지는 경우로, 노드가 일렬로 늘어선 형태이다.

- **최대 높이 = n** (노드 수와 동일)

### 최소 높이
이진 검색 트리가 **완전 이진 트리**에 가까운 형태일 때 높이가 최소이다.  
높이 h인 완전 이진 트리의 노드 수: 2^(h-1) ≤ n ≤ 2^h - 1  
→ h ≥ log₂(n+1)  

- **최소 높이 = ⌈log₂(n+1)⌉**

**정답:**

 최대 높이  n (편향 이진 트리)

 최소 높이  log₂(n+1) (완전 이진 트리)
