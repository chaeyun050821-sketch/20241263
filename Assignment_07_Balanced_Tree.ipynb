## 문제 01
**이진 검색 트리에서 균형이 맞지 않으면, 최악의 경우 점근적 검색 시간과 최선의 경우 점근적 검색 시간은 어떻게 되는가?**

### 풀이

이진 검색 트리(BST)에서 균형이 맞지 않을 경우:

| 경우 | 시간 복잡도 | 설명 |
|------|------------|------|
| **최악의 경우 (Worst Case)** | **O(n)** | 트리가 한쪽 방향으로만 치우친 경우 (예: 오름차순으로 삽입 시 선형 구조가 됨). 이 경우 탐색은 연결 리스트 순차 탐색과 동일 |
| **최선의 경우 (Best Case)** | **O(log n)** | 균형이 깨졌어도 찾는 키가 루트에 있는 경우. 단, 이는 특정 탐색 하나에 대한 최선이며, 트리 전체 구조상 최선은 여전히 O(log n) |

**결론:**
- 최악의 경우 점근적 검색 시간: **Θ(n)**
- 최선의 경우 점근적 검색 시간: **Θ(1)** (루트가 탐색 대상인 경우) 또는 트리 탐색 관점에서 **Θ(log n)**

> 균형이 깨진 BST의 높이는 최대 n이 되므로, 탐색 시간도 최대 O(n)이 된다.

## 문제 02
**아래 AVL 트리의 균형이 깨진 상태는 삽입과 삭제 중 어느 한쪽의 작업을 한 후인가, 아니면 둘 다 가능한가?**

```
        40
       /  \
      10   55
        \    \
        20   70
       /  \
      17  30
```

### 풀이

각 노드의 균형 인수(Balance Factor = 왼쪽 서브트리 높이 - 오른쪽 서브트리 높이)를 계산한다.

```
노드별 높이 계산:
- 17: 높이 0, BF = 0
- 30: 높이 0, BF = 0
- 20: 높이 1, BF = 0 (양쪽 자식 있음)
- 70: 높이 0, BF = 0
- 10: 높이 2, BF = -2  ← 균형 위반! (왼쪽 없음, 오른쪽 높이 2)
- 55: 높이 1, BF = -1
- 40: 높이 3, BF = 0 (왼쪽 높이 2, 오른쪽 높이 1) → BF = 2-1 = 1 ... 확인 필요
```

**상세 계산:**
- leaf: 17(h=1), 30(h=1), 70(h=1)
- 20: left=17(h=1), right=30(h=1) → h=2, BF=0
- 10: left=없음(h=0), right=20(h=2) → h=3, BF = 0-2 = **-2** ← **균형 위반**
- 55: left=없음(h=0), right=70(h=1) → h=2, BF = 0-1 = -1
- 40: left=10(h=3), right=55(h=2) → h=4, BF = 3-2 = 1

**답: 삽입과 삭제 둘 다 가능하다.**

- **삽입으로 발생 가능:** 10의 왼쪽 자식이 없는 상태에서 20 아래에 17 또는 30을 삽입하면 10의 BF가 -2가 됨
- **삭제로 발생 가능:** 원래 10의 왼쪽에 노드가 있었는데 삭제하면 BF가 -2가 됨

→ **삽입과 삭제 모두 이 불균형을 유발할 수 있다.**

## 문제 03
**아래 AVL 트리의 균형이 깨진 상태는 삽입과 삭제 중 어느 한쪽의 작업을 한 후인가, 아니면 둘 다 가능한가?**

```
        40
       /  \
      10   55
     /  \
    5   20
        /
       17
```

### 풀이

각 노드의 균형 인수 계산:

```
- 5:  leaf → h=1, BF=0
- 17: leaf → h=1, BF=0
- 20: left=17(h=1), right=없음(h=0) → h=2, BF = 1-0 = +1
- 55: leaf → h=1, BF=0
- 10: left=5(h=1), right=20(h=2) → h=3, BF = 1-2 = -1
- 40: left=10(h=3), right=55(h=1) → h=4, BF = 3-1 = +2  ← 균형 위반!
```

40의 BF = +2로 균형이 깨짐. 40 기준 왼쪽이 더 깊고, 10 기준 오른쪽(BF=-1)이 더 깊음 → **LR 케이스**

**답: 삽입만 가능하다.**

**이유:**
- 불균형 노드(40)의 BF = +2, 자식(10)의 BF = -1
- 삭제 후 불균형이 생기는 경우, 불균형 노드의 더 깊은 쪽 자식(10)의 BF는 0이어야 하거나 같은 방향이어야 한다.
- 자식 10의 BF = -1 (LR 형태)은 삽입으로만 생성 가능: 10의 오른쪽 서브트리(20 아래)에 새 노드(17)를 삽입했을 때 이 구조가 만들어짐.
- 삭제의 경우 이 LR 패턴이 나타나려면 특수한 조건이 필요하며, 이 트리 구조는 **삽입에 의해서만 발생**한다.

> 구체적으로: 기존 AVL 트리에서 17을 삽입하면 40→10→20→17 경로로 인해 40의 BF가 +2가 되어 이 불균형 상태가 됨.

## 문제 04
**03번 문제의 트리가 AVL 트리가 되도록 수선하고, 수선 과정을 모두 보여라.**

### 풀이

불균형 상태:
```
        40  (BF=+2)
       /  \
     10    55
    /  \
   5   20  (BF=+1, 왼쪽이 더 깊음)
      /
     17
```

**불균형 분석:**
- 불균형 노드: **40** (BF = +2, 왼쪽이 무거움)
- 왼쪽 자식 10의 BF = -1 (오른쪽이 무거움)
- → **LR (Left-Right) 케이스**

**LR 케이스 수선 방법: 2번의 회전 필요**

---

**Step 1: 10을 기준으로 Left Rotation (왼쪽 회전)**

10의 오른쪽 자식인 20을 중심으로 Left Rotate:

```
        40  (BF=+2)
       /  \
     20    55
    /  \
   10  (없음)
  /  \
 5   17
```

즉:
- 20이 10의 자리로 올라옴
- 10은 20의 왼쪽 자식
- 17(20의 왼쪽 자식이었던)은 10의 오른쪽 자식

```
        40
       /  \
      20   55
     /
    10
   /  \
  5   17
```

---

**Step 2: 40을 기준으로 Right Rotation (오른쪽 회전)**

40의 왼쪽 자식 20을 중심으로 Right Rotate:

- 20이 루트로 올라옴
- 40은 20의 오른쪽 자식
- 20의 오른쪽 자식(없음)은 40의 왼쪽 자식

```
        20
       /  \
      10   40
     /  \    \
    5   17   55
```

---

**결과 검증 (각 노드의 BF 확인):**

| 노드 | 왼쪽 높이 | 오른쪽 높이 | BF | 균형? |
|------|-----------|-------------|-----|-------|
| 5    | 0         | 0           | 0   | ✅    |
| 17   | 0         | 0           | 0   | ✅    |
| 55   | 0         | 0           | 0   | ✅    |
| 10   | 1         | 1           | 0   | ✅    |
| 40   | 0         | 1           | -1  | ✅    |
| 20   | 2         | 2           | 0   | ✅    |

**모든 노드의 |BF| ≤ 1 → AVL 트리 성질 만족!**

## 문제 05
**AVL 트리는 균형이 깨진 스타일을 LL, LR, RR, RL의 네 가지 케이스 중 하나로 처리한다. 이들 중 한 번의 회전으로 균형이 맞추어지지 않는 경우는 어느 것인가?**

### 풀이

**답: LR 케이스와 RL 케이스는 한 번의 회전으로 균형이 맞지 않는다.**

각 케이스별 필요 회전 수:

| 케이스 | 회전 방향 | 회전 횟수 | 설명 |
|--------|-----------|-----------|------|
| **LL** | Right Rotation 1회 | **1회** | 불균형 노드 기준 오른쪽 회전 1번으로 해결 |
| **RR** | Left Rotation 1회  | **1회** | 불균형 노드 기준 왼쪽 회전 1번으로 해결 |
| **LR** | Left → Right 회전  | **2회** | 자식 노드 Left Rotation 후 불균형 노드 Right Rotation |
| **RL** | Right → Left 회전  | **2회** | 자식 노드 Right Rotation 후 불균형 노드 Left Rotation |

**이유:**
- LL/RR은 불균형이 한쪽 방향으로 일관되게 쏠려 있어 단일 회전으로 해결됨
- LR/RL은 불균형 방향이 "꺾여" 있어 (left-right 또는 right-left), 먼저 직선 형태로 바꾸는 1차 회전 후 다시 단일 회전이 필요함

```
LL 예:          RL 예 (2회 필요):
    C              C
   /              /
  B    →(1회)   B      →(꺾인 형태, 2회 필요)
 /               \
A                 A
```

## 문제 06
**다음은 맞는 주장인가?**

> *AVL 트리에서 균형이 맞지 않아 수선하고 난 후 수선 전에 비해 AVL 트리의 높이가 반드시 낮아진다.*

### 풀이

**답: 틀린 주장이다.**

**이유:**

수선(회전) 후 AVL 트리의 높이 변화는 **삽입인지 삭제인지에 따라 다르다.**

| 연산 | 수선 후 높이 변화 |
|------|------------------|
| **삽입 후 수선** | 수선 전 불균형 서브트리의 높이와 **동일하게 유지됨** (낮아지는 게 아니라 삽입 전 높이로 복원됨) |
| **삭제 후 수선** | 수선 후 높이가 **낮아질 수 있음** |

**삽입의 경우 상세:**
- 삽입 전 서브트리 높이: h
- 삽입 후 수선 전: h+1 (불균형)
- 수선(회전) 후: h (원래 높이로 복원, 낮아진 게 아님)

따라서 "반드시 낮아진다"는 주장은 **틀렸다.** 삽입 후 수선의 경우 높이는 삽입 *직전* 대비 변화가 없으며 (삽입으로 높아졌다가 수선으로 되돌아옴), 삭제 후 수선의 경우에만 실질적으로 낮아질 수 있다.

## 문제 07
**네 가지 타입에 맞추어 AVL 트리를 수선한 후 여전히 균형이 맞지 않아 부모 노드에 대해 수선해줘야 하는 경우가 있다. 이렇게 위로 올라가면서 3회 이상 수선이 필요한 경우가 발생할 수 있는가?**

### 풀이

**답: 삽입의 경우 최대 1번, 삭제의 경우 O(log n)번 수선이 필요할 수 있다.**

**삽입의 경우:**
- 삽입 후 수선은 **최대 1회**의 회전(또는 이중 회전)으로 완료된다.
- 수선이 완료되면 해당 서브트리의 높이가 삽입 전과 동일하게 복원되므로, 상위 노드들의 BF도 변하지 않아 추가 수선이 불필요하다.
- 따라서 삽입에서 3회 이상 수선은 **불가능하다.**

**삭제의 경우:**
- 삭제 후 수선은 **최악의 경우 O(log n)번** 필요할 수 있다.
- 한 노드에서 수선(회전)을 해도 그 서브트리의 높이가 줄어들 수 있고, 그로 인해 부모 노드에서 다시 불균형이 발생할 수 있다.
- 이 과정이 루트까지 전파될 수 있으므로, 3회 이상은 물론 O(log n)회까지 수선이 필요한 경우가 **발생할 수 있다.**

**결론:** 
- 삽입 → 최대 1회 수선 (3회 이상 불가)
- 삭제 → 3회 이상 수선 **가능**, 최악 O(log n)회

## 문제 08
**다음 RB 트리에서 키 7이 삽입되고 균형을 맞춘 후의 모양과 노드의 색상을 그리시오.**

초기 RB 트리:
```
        10(B)
       /     \
     5(R)   13(B)
    /   \       \
  3(B)  6(B)   15(R)
  /       \
2(R)      8(R)
```

### 풀이

**Step 1: 키 7 삽입 위치 찾기**

BST 삽입 규칙: 10 → 5 → 6 → 8 → 7은 8의 왼쪽

새 노드 7을 **Red**로 삽입:
```
        10(B)
       /     \
     5(R)   13(B)
    /   \       \
  3(B)  6(B)   15(R)
  /       \
2(R)      8(R)
          /
         7(R)  ← 삽입
```

**Step 2: RB 트리 성질 확인**

- **성질 3 위반**: 8(R)의 자식 7(R) → Red-Red 연속 (Double Red 발생!)

**Step 3: 수선 - Uncle 노드 확인**

- 삽입 노드 7의 부모: 8(R)
- 부모 8의 부모(조부모): 6(B)
- Uncle 노드: 6의 왼쪽 자식 = **없음(Black, NIL)**

Uncle이 Black이므로 → **회전으로 수선**

형태 분석: 6 → 8(오른쪽) → 7(왼쪽) = **RL 케이스**

**Step 4: RL 회전 수행**

*4-1. 8을 기준으로 Right Rotation:*
```
  6(B)
    \
     7(R)
       \
        8(R)
```

*4-2. 6을 기준으로 Left Rotation + 색상 조정 (7↔6 색상 교환):*
```
  7(B)
 /   \
6(R)  8(R)
```

**Step 5: 전체 트리 최종 모양**

```
           10(B)
          /     \
        5(R)   13(B)
       /   \       \
     3(B)  7(B)   15(R)
     /    /   \
   2(R) 6(R)  8(R)
```

**RB 트리 성질 검증:**
- ✅ 루트(10)는 Black
- ✅ Red 노드의 자식은 모두 Black (Double Red 없음)
- ✅ 각 경로의 Black 노드 수 동일 (Black-height = 2)
- ✅ 모든 성질 만족

## 문제 09
**RB트리에서 리프 노드에 새 키를 삽입하고 레드를 칠하니 성질3이 깨졌다. 이를 수선하는 과정에서 최악의 경우에는 루트까지 영향을 미칠 수도 있다. 루트까지 영향을 미치는 상황을 설명하고, 삽입이 일어난 노드를 p라 할 때 적어도 p⁶까지는 어떤 일이 일어나는지 보여라.**

### 풀이

**핵심: Double Red가 루트 방향으로 계속 전파되는 시나리오**

Double Red 수선 시 Uncle이 **Red**이면 색상 재칠(Recoloring)을 하고 문제가 위로 전파된다.

---

**전파 조건 (매 단계):**
- 현재 Double Red 발생 위치의 Uncle 노드가 **Red**인 경우
- → 부모와 삼촌을 Black, 조부모를 Red로 재칠
- → 조부모가 새로운 Red 노드가 되어 그 부모와 Double Red 가능

---

**p에서 p⁶까지의 과정 (모든 Uncle이 Red인 최악의 경우):**

```
초기 삽입:
  p   = 삽입 노드 (Red)
  p¹  = p의 부모 (Red) → Double Red 발생!
  p²  = p의 조부모 (Black), Uncle(p¹의 형제) = Red

[Case: Uncle이 Red → Recoloring]
  p¹ → Black으로 변경
  Uncle → Black으로 변경  
  p² → Red로 변경
  → p²와 p³ 사이에서 새로운 Double Red 발생!

  p²  = 이제 Red
  p³  = p²의 부모 (Red) → Double Red!
  p⁴  = p³의 부모 (Black), Uncle(p³의 형제) = Red

[다시 Recoloring]
  p³ → Black
  Uncle → Black
  p⁴ → Red
  → p⁴와 p⁵ 사이에서 Double Red!

  p⁴  = Red
  p⁵  = p⁴의 부모 (Red) → Double Red!
  p⁶  = p⁵의 부모 (Black), Uncle(p⁵의 형제) = Red

[다시 Recoloring 또는 회전]
  Uncle이 Red면 → p⁵ Black, Uncle Black, p⁶ Red → 위로 계속 전파
  Uncle이 Black이면 → 회전으로 최종 해결 (전파 종료)
```

---

**단계별 요약표:**

| 단계 | 노드 | 역할 | 처리 |
|------|------|------|------|
| p | 삽입 노드 | Red (삽입) | Double Red 시작 |
| p¹ | 부모 | Red | Double Red 발생 |
| p² | 조부모 | Black → **Red** | Recoloring (Uncle이 Red인 경우) |
| p³ | 증조부모 | Red | p²와 Double Red 재발생 |
| p⁴ | 고조부모 | Black → **Red** | Recoloring |
| p⁵ | 고고조부모 | Red | p⁴와 Double Red 재발생 |
| p⁶ | 고고조할머니 | Black → **Red** | Recoloring 또는 회전으로 해결 |

---

**결론:**
- Uncle이 Red인 경우가 계속 반복되면 Double Red가 2칸씩 위로 전파됨
- 최악의 경우 루트까지 도달하며, 루트가 Red가 되면 다시 Black으로 칠하여 종료

## 문제 14
**B-트리에서 리프가 아닌 노드에 있는 원소의 직후 원소는 항상 리프 노드에 있다. 왜 그런가?**

### 풀이

**답:**

B-트리의 특성상, 내부 노드의 키 k에 대해 직후 원소(Inorder Successor)는 k의 오른쪽 서브트리에서 가장 작은 값이다.

**B-트리의 구조적 특성:**
- B-트리의 모든 리프 노드는 같은 레벨(깊이)에 있다.
- 내부 노드의 오른쪽 서브트리의 "가장 왼쪽 경로"를 따라 내려가면, 결국 리프 노드에 도달한다.

**귀납적 설명:**
1. 내부 노드의 키 k의 오른쪽 포인터가 가리키는 자식 노드 C가 있다.
2. C가 내부 노드라면, C의 첫 번째 키는 k보다 크지만, C의 왼쪽 포인터가 가리키는 자식에는 더 작은 값(하지만 k보다 큰)이 있다.
3. 이를 반복하면 결국 더 이상 내려갈 자식이 없는 **리프 노드**에 도달한다.
4. 그 리프 노드의 첫 번째 키가 k의 직후 원소다.

**핵심 이유:** B-트리에서 실제 데이터의 순서 관계는 리프 레벨에서 완성되며, 내부 노드의 키는 서브트리를 분리하는 "분리자(separator)" 역할을 하기 때문에, 어떤 키의 직후 원소는 반드시 그 키의 오른쪽 서브트리의 최솟값이고, 이는 항상 리프 노드에 위치한다.

## 문제 15
**디스크의 한 블록이 8,192바이트, 키의 크기가 20바이트, 페이지 번호가 4바이트를 차지한다면, B-트리의 한 노드는 최대 몇 개의 키를 가질 수 있도록 설계하겠는가?**

````python
# B-트리 노드의 최대 키 수 계산

block_size = 8192   # 바이트
key_size = 20       # 바이트
page_size = 4       # 바이트 (포인터/페이지 번호)

# B-트리 노드 구조:
# k개의 키와 (k+1)개의 자식 포인터를 저장
# 크기: k * key_size + (k+1) * page_size <= block_size
# k * 20 + (k+1) * 4 <= 8192
# 20k + 4k + 4 <= 8192
# 24k <= 8188
# k <= 341.16...

k = (block_size - page_size) // (key_size + page_size)
print(f"블록 크기: {block_size} bytes")
print(f"키 크기: {key_size} bytes")
print(f"페이지 번호 크기: {page_size} bytes")
print()
print("노드 구조: k개의 키 + (k+1)개의 자식 포인터")
print(f"공식: k × {key_size} + (k+1) × {page_size} ≤ {block_size}")
print(f"      {key_size+page_size}k + {page_size} ≤ {block_size}")
print(f"      k ≤ ({block_size} - {page_size}) / {key_size+page_size} = {(block_size-page_size)/(key_size+page_size):.4f}")
print()
print(f"최대 키 수 k = {k}")
print(f"검증: {k} × {key_size} + ({k+1}) × {page_size} = {k*key_size + (k+1)*page_size} ≤ {block_size} ✅")
````

### 풀이 정리

B-트리 노드는 **k개의 키**와 **k+1개의 자식 포인터**를 저장한다.

$$k \times 20 + (k+1) \times 4 \leq 8192$$
$$24k + 4 \leq 8192$$
$$24k \leq 8188$$
$$k \leq 341.16...$$

**따라서 한 노드의 최대 키 수는 341개로 설계한다.**

검증: 341 × 20 + 342 × 4 = 6,820 + 1,368 = 8,188 ≤ 8,192 ✅

## 문제 16
**각 노드가 최대 k개의 키를 가질 수 있는 B-트리에서 리프 노드에 들어 있는 키의 수와 내부 노드에 들어 있는 키의 수는 어떤 관계가 있는가?**

### 풀이

**결론: 리프 노드의 총 키 수 = 내부 노드의 총 키 수 + 1**

**증명:**

B-트리에서 내부 노드는 자식들 사이의 "분리자(separator)" 역할을 한다.

- 내부 노드가 k개의 키를 가지면 k+1개의 자식 포인터를 가진다.
- 전체 트리의 포인터(자식 링크) 수 = 내부 노드의 총 자식 수
- 전체 트리의 노드 수 = 리프 수 + 내부 노드 수

n개의 내부 노드가 있고, 각각 자식이 있다고 하면:

$$\text{리프 수} = \text{내부 노드의 총 자식 포인터 수} - \text{내부 노드 수} + 1$$

좀 더 직관적으로:

| 구분 | 관계 |
|------|------|
| 내부 노드 키가 m개 | 내부 노드의 자식이 m+1개 |
| 리프 노드 키의 총합 | (내부 노드 총 키 수 + 1)개의 "구간"을 분할 |

**실용적 의미:**
- 리프 노드의 총 키 수 ≈ 내부 노드의 총 키 수 × (평균 분기 수 - 1)
- 실제 데이터(키)의 대부분은 리프 노드에 분포
- 내부 노드 키 수 << 리프 노드 키 수 (차이가 매우 큼)

예: 각 노드에 최대 k=5개의 키, 리프 레벨에 n개의 키가 있다면 내부 노드 수는 약 n/5 × (1/5) + ... ≈ n/4 정도

## 문제 17
**B-트리에서 언더플로우가 일어나 자신의 형제 노드와 병합할 때 두 노드의 키들을 모두 합치고 부모 노드에서 추가로 1개의 키를 가져온다. 이렇게 다 합치면 병합으로 만들어진 노드에서 오버플로우가 생기는 경우는 없는가?**

### 풀이

**답: 오버플로우는 절대 발생하지 않는다.**

**수학적 증명:**

최대 키 수를 k라 하면, 최소 키 수는 ⌈k/2⌉ - 1이다.

병합이 발생하는 조건:
- 언더플로우 노드의 키 수: **최소 0개** (삭제 후), 일반적으로 ⌈k/2⌉ - 2개
- 형제 노드의 키 수: 빌려줄 수 없으므로 **정확히 ⌈k/2⌉ - 1개** (최솟값)
- 부모에서 가져오는 키: **1개**

병합 후 키 수 합계 (최악의 경우):

$$\text{병합 후 키 수} = (⌈k/2⌉ - 2) + (⌈k/2⌉ - 1) + 1$$
$$= 2 \times ⌈k/2⌉ - 2$$

**k가 짝수일 때:** ⌈k/2⌉ = k/2 → 2 × k/2 - 2 = k - 2 < k ✅  
**k가 홀수일 때:** ⌈k/2⌉ = (k+1)/2 → 2 × (k+1)/2 - 2 = k + 1 - 2 = k - 1 < k ✅

어느 경우에도 병합 후 키 수 ≤ k - 1 < k 이므로, **오버플로우(키 수 > k)는 절대 발생하지 않는다.**

**직관적 이해:**
- 병합 조건 자체가 "형제가 최솟값의 키만 가지고 있어 빌려줄 수 없는 경우"이므로,
- 두 노드의 키를 모두 합쳐도 최대 k개 이하가 되도록 B-트리의 최솟값이 설계되어 있다.

## 문제 18
**B-트리의 각 노드가 담을 수 있는 최대 키 수가 1,000개인 예에서 1조 개의 키로 인덱싱을 하면 B-트리의 높이의 범위는 어떻게 되는가? (하나의 노드로 이루어진 트리의 높이를 1로 정의)**

````python
import math

k = 1000          # 최대 키 수
n = 10**12        # 1조 개의 키

# 최솟값: 최소 키 수
min_keys_per_node = math.ceil(k / 2) - 1  # = 499
min_children = math.ceil(k / 2)           # = 500 (루트 제외 내부 노드의 최소 자식 수)

print(f"최대 키 수 (k): {k}")
print(f"최소 키 수: {min_keys_per_node}")
print(f"최소 분기 수: {min_children}")
print(f"최대 분기 수: {k + 1}")
print(f"총 키 수 (n): {n:,}")
print()

# 최소 높이: 모든 노드가 꽉 찬 경우 (최대 분기)
# 높이 h인 트리에 저장 가능한 최대 키 수
# 루트: k개, 2레벨: (k+1)*k개, ... 
# 리프 수 최대 = (k+1)^(h-1)
# 총 키 최대 = k * ((k+1)^h - 1) / k ... 복잡하므로 리프 기준으로 계산
# 리프 수 최대일 때 총 키 최대
# 최소 높이: ceil(log_{k+1}(n+1)) 근사

min_height = math.ceil(math.log(n + 1, k + 1))
print(f"최소 높이 (모든 노드 가득 찬 경우):")
print(f"  h_min ≈ ⌈log_{k+1}({n:,})⌉ = ⌈{math.log(n, k+1):.4f}⌉ = {min_height}")
print()

# 최대 높이: 모든 노드가 최솟값의 키만 가진 경우
# 루트는 최소 1개의 키 (자식 2개)
# 나머지 내부 노드는 최소 ceil(k/2)-1개 키, ceil(k/2)개 자식
# 리프 수 최소 = 2 * ceil(k/2)^(h-2) for h>=2
# n <= 리프수 * k, 이를 역산
# 최대 높이: 리프 노드 수가 최소일 때
# 최소 리프 수: 2 * (⌈k/2⌉)^(h-1)
# n <= k * 2 * (⌈k/2⌉)^(h-1)
# 역산: h-1 >= log_{⌈k/2⌉}(n / (2k))

max_height = math.ceil(math.log(n / 2, min_children)) + 1
print(f"최대 높이 (모든 노드 최솟값인 경우):")
print(f"  h_max ≈ ⌈log_{min_children}(n/2)⌉ + 1")
print(f"        = ⌈log_{min_children}({n//2:,})⌉ + 1")
print(f"        = ⌈{math.log(n/2, min_children):.4f}⌉ + 1")
print(f"        = {max_height}")
print()
print(f"결론: B-트리의 높이 범위 = {min_height} ≤ h ≤ {max_height}")
````

### 풀이 정리

**주어진 조건:** k = 1,000 (최대 키 수), n = 10¹² (1조)

**최소 높이 (모든 노드가 최대 키 수로 꽉 찬 경우):**
$$h_{min} = \lceil \log_{1001}(10^{12}) \rceil = \lceil 4.0 \rceil = 4$$

**최대 높이 (모든 노드가 최솟값의 키만 가진 경우):**
- 최소 분기 수: ⌈1000/2⌉ = 500
- 루트는 최소 1개의 키(자식 2개)

$$h_{max} = \lceil \log_{500}(10^{12}/2) \rceil + 1 \approx 5$$

**결론: B-트리의 높이 범위는 4 ≤ h ≤ 5**

> B-트리의 장점: 1조 개의 키를 겨우 4~5번의 디스크 접근으로 탐색 가능

## 문제 19
**B-트리 인덱스는 기본적으로 디스크에 인덱스가 저장된 상태로 검색하는 것을 전제로 한다. 현장에서 사용할 때 트리의 모양을 바꾸지 않으면서 실용적으로 디스크 접근 횟수를 줄일 수 있는 아이디어를 말해보시오.**

### 풀이

트리 모양을 바꾸지 않으면서 디스크 접근 횟수를 줄이는 방법:

---

**1. 루트 노드(및 상위 레벨 노드) 메모리 캐싱**

- B-트리의 루트는 **항상** 탐색 경로에 포함된다.
- 상위 1~2 레벨 노드들도 대부분의 탐색에서 접근된다.
- → **루트와 상위 레벨 노드들을 메인 메모리(RAM)에 항상 캐싱**해두면 해당 레벨의 디스크 I/O를 완전히 제거할 수 있다.
- 예: 높이 5인 B-트리에서 상위 2레벨 캐싱 시 디스크 접근 횟수: 5 → 3 (40% 절감)

---

**2. 버퍼 풀(Buffer Pool) / 페이지 캐시 활용**

- 최근 또는 자주 접근한 노드(디스크 블록)를 메모리에 버퍼링 (LRU 등 교체 정책 사용)
- 반복 탐색 시 캐시 히트로 디스크 접근 생략
- 대부분의 RDBMS(MySQL InnoDB, PostgreSQL 등)에서 실제로 사용하는 방법

---

**3. 프리페칭(Prefetching)**

- 탐색 패턴이 예측 가능한 경우(범위 검색 등), 다음에 접근할 노드를 미리 디스크에서 읽어 메모리에 적재
- 실제 접근 시점에는 디스크 I/O가 이미 완료되어 있어 대기 시간 감소

---

**4. 운영체제의 파일 시스템 캐시 활용**

- OS 레벨의 페이지 캐시를 이용하여 자주 접근되는 B-트리 노드들이 자동으로 메모리에 유지되도록 설정

---

**핵심 아이디어 요약:**

| 방법 | 원리 | 효과 |
|------|------|------|
| 루트/상위 노드 캐싱 | 자주 접근되는 노드 고정 | 직접적 I/O 제거 |
| 버퍼 풀 | LRU 기반 캐시 | 반복 접근 I/O 제거 |
| 프리페칭 | 비동기 선행 로드 | 대기 시간 감소 |

> 가장 효과적인 방법: **루트를 포함한 상위 레벨 노드를 서버 기동 시 메모리에 고정**. 1조 개 키, 높이 5인 B-트리에서 상위 2레벨만 캐싱해도 대부분의 탐색에서 디스크 접근 3회로 줄일 수 있다.
