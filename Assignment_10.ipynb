#### 16. 다음 그래프에서 위상 정렬을 할 때, STACK을 사용한 위상 정렬 수도 코드를 쓰고, 가능한 위상 정렬 순서가 모두 몇 가지인지 쓰라.

```python
TopologicalSort(Graph G):
    Stack S  // 진입 차수가 0인 정점을 담을 스택
    배열 in_degree[G.V] // 각 정점의 진입 차수를 저장하는 배열

    // 1. 초기화: 모든 정점의 진입 차수를 계산
    for each u in G.V:
        in_degree[u] = u로 들어오는 간선의 수

    // 2. 진입 차수가 0인 시작 정점들을 스택에 삽입
    for each u in G.V:
        if in_degree[u] == 0:
            S.push(u)

    // 3. 스택이 빌 때까지 반복하며 정렬 수행
    while S is not empty:
        u = S.pop()
        print(u)  // 정점 출력 (위상 정렬 결과)

        // u와 연결된 인접 정점들의 진입 차수를 1씩 감소
        for each v in u.adj:
            in_degree[v] = in_degree[v] - 1
            // 진입 차수가 0이 되면 스택에 삽입
            if in_degree[v] == 0:
                S.push(v)
```

시작점 고정: 진입 차수가 0인 정점은 a뿐이므로, 무조건 a가 첫 번째로 와야함.
두 갈래의 독립된 경로: a 이후로 그래프는 서로 영향을 주지 않는 두 개의 그룹으로 나뉨.    

왼쪽 그룹 (2개 노드): b → c     
오른쪽 그룹 (5개 노드): d → e → f → (g, h)   

간선들을 자세히 보면 f는 d와 e를 모두 거쳐야 하고, g는 e와 f를 모두 거쳐야 한다. 즉, d → e → f의 순서는 절대적으로 고정된다.f 이후에 g와 h가 올 수 있는데, g와 h 사이에는 선후 관계가 없기 때문에 오른쪽 그룹 내에서 가능한 순서는 defgh 와 defhg 딱 2가지뿐이다.   

경우의 수 조합 (Interleaving):이제 첫 글자 a를 제외한 나머지 7개의 자리에, 왼쪽 그룹(2개)과 오른쪽 그룹(5개)을 순서 규칙을 유지한 채 섞어 넣는다.총 7개의 빈자리 중 왼쪽 그룹 b, c가 들어갈 2개의 자리를 고르는 조합의 수는 $_7C_2 = (7 \times 6) / 2 = 21$가지이다. 여기에 오른쪽 그룹 자체가 내부적으로 가질 수 있는 순서 2가지를 곱해준다.최종 계산: $21 \times 2 = \mathbf{42}$

* **4회전:** 나머지 중 최솟값 30을 선택. 네 번째 자리(70)와 교환
  $\rightarrow$ `[12, 20, 25, 30, 55, 70]` (30 정렬 완료)
* **5회전:** 나머지 중 최솟값 55를 선택. 다섯 번째 자리(55)와 교환 (변화 없음)
  $\rightarrow$ `[12, 20, 25, 30, 55, 70]` (정렬 완료)

#### (2) 버블 정렬 (Bubble Sort) 수행 과정
*인접한 두 원소를 비교하여 큰 값을 뒤로 보내는 과정을 반복합니다.*

* **초기 상태:** `[12, 70, 30, 20, 55, 25]`
* **1회전:** (70,30), (70,20), (70,55), (70,25) 차례로 교환 발생. 가장 큰 70이 맨 뒤로 이동
  $\rightarrow$ `[12, 30, 20, 55, 25, 70]` (70 정렬 완료)
* **2회전:** (30,20), (55,25) 교환 발생. 두 번째로 큰 55가 뒤로 이동
  $\rightarrow$ `[12, 20, 30, 25, 55, 70]` (55 정렬 완료)
* **3회전:** (30,25) 교환 발생.
  $\rightarrow$ `[12, 20, 25, 30, 55, 70]` (30 정렬 완료)
* **4회전:** 인접한 원소들(12-20, 20-25) 비교 시 모두 정렬되어 있어 교환 없음.
  $\rightarrow$ `[12, 20, 25, 30, 55, 70]` (정렬 완료)

#### (3) 삽입 정렬 (Insertion Sort) 수행 과정
*두 번째 원소부터 시작하여, 그 앞의 정렬된 부분과 비교해 알맞은 위치에 삽입합니다.*

* **초기 상태:** `[12, 70, 30, 20, 55, 25]` (첫 번째 원소 12는 정렬된 것으로 간주)
* **1회전 (70 삽입):** 70은 12보다 크므로 제자리에 위치
  $\rightarrow$ `[12, 70, 30, 20, 55, 25]`
* **2회전 (30 삽입):** 30을 70 앞, 12 뒤에 삽입 (70은 한 칸 뒤로 밀림)
  $\rightarrow$ `[12, 30, 70, 20, 55, 25]`
* **3회전 (20 삽입):** 20을 30 앞, 12 뒤에 삽입 (30, 70은 뒤로 밀림)
  $\rightarrow$ `[12, 20, 30, 70, 55, 25]`
* **4회전 (55 삽입):** 55를 70 앞, 30 뒤에 삽입 (70은 뒤로 밀림)
  $\rightarrow$ `[12, 20, 30, 55, 70, 25]`
* **5회전 (25 삽입):** 25를 30 앞, 20 뒤에 삽입 (30, 55, 70은 뒤로 밀림)
  $\rightarrow$ `[12, 20, 25, 30, 55, 70]` (정렬 완료)

#### (1) 병합 정렬 (Merge Sort) 수행 과정
*배열을 크기가 1이 될 때까지 반으로 나눈 후, 정렬하면서 다시 합치는(Merge) 방식입니다.*

* **[분할 단계]**
  * **초기:** `[12, 70, 30, 20, 55, 25, 40, 50]`
  * **1차 분할:** `[12, 70, 30, 20]` / `[55, 25, 40, 50]`
  * **2차 분할:** `[12, 70]` / `[30, 20]` / `[55, 25]` / `[40, 50]`
  * **3차 분할:** `[12]`, `[70]`, `[30]`, `[20]`, `[55]`, `[25]`, `[40]`, `[50]` (모두 크기 1로 분할 완료)

* **[병합 단계]**
  * **1차 병합:** `[12, 70]` / `[20, 30]` / `[25, 55]` / `[40, 50]` (크기 2로 병합하며 정렬)
  * **2차 병합:** `[12, 20, 30, 70]` / `[25, 40, 50, 55]` (크기 4로 병합하며 정렬)
  * **3차 병합:** `[12, 20, 25, 30, 40, 50, 55, 70]` (최종 병합 및 정렬 완료)

#### (2) 힙 정렬 (Heap Sort) 수행 과정
*배열을 최대 힙(Max Heap) 트리 구조로 만든 뒤, 루트(최댓값)를 뽑아내어 맨 뒤로 보내는 과정을 반복합니다.*

* **1단계: 최대 힙(Max Heap) 구성**
  * 초기 배열: `[12, 70, 30, 20, 55, 25, 40, 50]`
  * 아래에서 위로(Bottom-up) 자식 노드와 비교하며 힙 생성
  * 노드 20 조정: 50과 교환 `[... 30, 50, 55, 25, 40, 20]`
  * 노드 30 조정: 40과 교환 `[... 40, 50, 55, 25, 30, 20]`
  * 노드 12(루트) 조정: 70, 55와 연속 교환
  * **최대 힙 완성:** `[70, 55, 40, 50, 12, 25, 30, 20]`

* **2단계: 최댓값 추출 및 정렬 (루트를 맨 뒤로 보내고 다시 힙 구성)**
  * **1회전:** 70 추출, `[55, 50, 40, 20, 12, 25, 30]` | 정렬: `[70]`
  * **2회전:** 55 추출, `[50, 30, 40, 20, 12, 25]` | 정렬: `[55, 70]`
  * **3회전:** 50 추출, `[40, 30, 25, 20, 12]` | 정렬: `[50, 55, 70]`
  * **4회전:** 40 추출, `[30, 20, 25, 12]` | 정렬: `[40, 50, 55, 70]`
  * **5회전:** 30 추출, `[25, 20, 12]` | 정렬: `[30, 40, 50, 55, 70]`
  * **6회전:** 25 추출, `[20, 12]` | 정렬: `[25, 30, 40, 50, 55, 70]`
  * **7회전:** 20 추출, `[12]` | 정렬: `[20, 25, 30, 40, 50, 55, 70]`
  * **최종 결과:** `[12, 20, 25, 30, 40, 50, 55, 70]`

#### (3) 퀵 정렬 (Quick Sort) 수행 과정
*기준값(Pivot)을 정해 그보다 작은 값은 왼쪽, 큰 값은 오른쪽으로 분할하는 과정을 반복합니다. (기준값은 가장 왼쪽 원소로 설정)*

* **초기 상태:** `[12, 70, 30, 20, 55, 25, 40, 50]`
* **1회전 (Pivot: 12):** 12보다 작은 값 없음, 모두 오른쪽으로 이동
  $\rightarrow$ `[12]` / `[70, 30, 20, 55, 25, 40, 50]` (12 정렬 완료)
* **2회전 (오른쪽 서브배열 Pivot: 70):** 70보다 큰 값 없음, 모두 왼쪽으로 이동
  $\rightarrow$ `[30, 20, 55, 25, 40, 50]` / `[70]` (70 정렬 완료)
* **3회전 (왼쪽 서브배열 Pivot: 30):** 30보다 작으면 왼쪽, 크면 오른쪽
  $\rightarrow$ `[20, 25]` / `[30]` / `[55, 40, 50]` (30 정렬 완료)
* **4회전 (서브배열들 정렬):**
  * `[20, 25]` 부분 (Pivot: 20) $\rightarrow$ `[20]` / `[25]`
  * `[55, 40, 50]` 부분 (Pivot: 55) $\rightarrow$ `[40, 50]` / `[55]`
* **5회전 (`[40, 50]` 정렬):**
  * Pivot: 40 $\rightarrow$ `[40]` / `[50]`
* **최종 병합 결과:** `[12, 20, 25, 30, 40, 50, 55, 70]` (정렬 완료)
